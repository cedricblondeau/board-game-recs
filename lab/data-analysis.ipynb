## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

## Load Kaggle data

```python
games = pd.read_csv("kaggle/games_detailed_info.csv")
games
```

```python
games.info()
```

```python
reviews = pd.read_csv("kaggle/bgg-19m-reviews.csv")
reviews
```

## Ratings distribution

```python
# Round ratings
reviews['rating'] = round(reviews['rating'], 0)
reviews["rating"].sort_values().unique()
```

```python
# Distribution of ratings
sns.countplot(x="rating", data=reviews, palette="viridis")
plt.title("Distribution of ratings", fontsize=14)
plt.show()
```

Most games receive between 6 and 9

## Games with the most reviews

```python
sns.countplot(
    x="name",
    data=reviews,
    palette="viridis",
    order=reviews.name.value_counts().iloc[:5].index
)
plt.title("Games with the most reviews", fontsize=8)
plt.show()

display(reviews.groupby('name').count()['rating'].sort_values(ascending=False).head(20))
```

Pandemic, Carcassonne and Catan are very popular with about more than 100k reviews each.

## Ratings per game

```python
reviews_grouped_by_game = reviews.groupby('name').count()
reviews_grouped_by_game = reviews_grouped_by_game.reset_index()[["name", "rating"]].rename(columns={"rating": "ratings_cnt"})

print(reviews_grouped_by_game['ratings_cnt'].mean())
print(reviews_grouped_by_game['ratings_cnt'].median())
print(reviews_grouped_by_game['ratings_cnt'].min())
print(reviews_grouped_by_game['ratings_cnt'].max())
```

A few learnings:
* On average, games have about 869 ratings.
* We have a median of 124 ratings per game.
* Game with the least ratings in the reviews dataset has 30 ratings.
* Game with the most ratings in the reviews dataset has 100k ratings (Pandemic).

```python
sns.histplot(data=reviews_grouped_by_game.ratings_cnt, stat="percent", bins=10)
```

However we can see that most games have only between 0 and a few thousands reviews, the most reviewed games are extreme outliers, let's zoom in.

```python
sns.histplot(data=reviews_grouped_by_game.ratings_cnt, stat="percent", bins=10, binrange=[100, 3000])
```
